```python
import re
import pandas as pd
import os
import numpy as np
```

```python
Data = pd.read_csv("DataL.csv")
Data = Data.dropna()
```

```python
entries = os.listdir('static/assets/2521')
```

```python
Data['SubImage'] ='good'
for i in range(len(Data)):
    for j in range(len(entries)):
        if Data["MainImage"][i][2:4]==entries[j][2:4]:
            Data['SubImage'][i]=entries[j]
        else:
            continue
```

```python
Data.head()
```

```python
for i in range(len(Data)):
    placefilternum = str(int(Data["PlaceFilter"].values[i]))
    entriesnum = str(int(entries[i][2:4]))
    for j in range(len(entries)):
        if placefilternum == entriesnum:
            Data["SubImage"][i].append(entries[j])
```
